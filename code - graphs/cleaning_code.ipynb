#### 2. Salary & Years of coding

```python
salary_yearscoding = copy.deepcopy(survey_results_2018)

salary_yearscoding = salary_yearscoding[['Salary','SalaryType' ,'YearsCoding','YearsCodingProf','Currency']]
salary_yearscoding = salary_yearscoding.dropna()
```

```python
#23_04_2024, 17:05 CEST
exchange_rates = {
    'British pounds sterling (£)': 1.24,
    'South African rands (R)': 0.052,
    'U.S. dollars ($)': 1,
    'Euros (€)': 1.07,
    'Swedish kroner (SEK)': 0.092,
    'Australian dollars (A$)': 0.65,
    'Indian rupees (₹)': 0.012,
    'Polish złoty (zł)': 0.25,
    'Russian rubles (₽)': 0.011,
    'Chinese yuan renminbi (¥)': 0.14,
    'Japanese yen (¥)': 0.0065,
    'Brazilian reais (R$)': 0.19,
    'Danish krone (kr)': 0.14,
    'Canadian dollars (C$)': 0.73,
    'Mexican pesos (MXN$)': 0.059,
    'Singapore dollars (S$)': 0.73,
    'Norwegian krone (kr)': 0.091 ,
    'Swiss francs': 1.10,
    'Bitcoin (btc)': 66958.60
}
```

```python
salary_yearscoding['Salary'] = salary_yearscoding['Salary'].str.replace(',', '')
salary_yearscoding['Salary'] = pd.to_numeric(salary_yearscoding['Salary'], errors='coerce', downcast='float')
salary_yearscoding = salary_yearscoding.dropna(subset=['Salary'])
```

```python
def convert_to_usd(row):
    if row['Currency'] in exchange_rates:
        if row['SalaryType'] == 'Monthly':
            return row['Salary'] * exchange_rates[row['Currency']] * 12
        if row['SalaryType'] == 'Weekly':
            return row['Salary'] * exchange_rates[row['Currency']] * 52
        return row['Salary'] * exchange_rates[row['Currency']]
    else:
        return float('NaN')

salary_yearscoding['SalaryUSD'] = salary_yearscoding.apply(convert_to_usd, axis=1)
salary_yearscoding = salary_yearscoding.dropna(subset=['SalaryUSD'])
```

```python
salary_yearscoding.to_csv('salary_yearscoding2018.csv')
salary_yearscoding
```

### 4. DevType vs OrgSize

```python
devtype_orgsize = copy.deepcopy(survey_results_2023)

devtype_orgsize = devtype_orgsize[['DevType', 'OrgSize']]

devtype_orgsize = devtype_orgsize.dropna()

grouped_dev_types = {
    'Developer': ['Developer, back-end', 'Developer, front-end', 'Developer, full-stack', 'Developer, desktop or enterprise applications',
                  'Developer, mobile', 'Developer, embedded applications or devices', 'Developer, game or graphics', 'Developer Experience',
                  'Developer Advocate'],
    'System_Admin': ['System administrator', 'Database administrator'],
    'Data_Science': ['Data scientist or machine learning specialist', 'Engineer, data', 'Research & Development role', 'Data or business analyst'],
    'Management': ['Senior Executive (C-Suite, VP, etc.)', 'Engineering manager', 'Product manager', 'Project manager'],
    'Other': ['Security professional', 'Designer', 'Hardware Engineer', 'Academic researcher', 'DevOps specialist', 'Engineer, site reliability',
              'Blockchain', 'Student', 'Educator', 'Scientist', 'Marketing or sales professional', 'Other (please specify):']
}


def assign_group(dev_type):
    for group, types in grouped_dev_types.items():
        if dev_type in types:
            return group
    return 'Other'

devtype_orgsize['Grouped_DevType'] = devtype_orgsize['DevType'].apply(assign_group)


org_size_order = ['Just me - I am a freelancer, sole proprietor, etc.', 
                  '2 to 9 employees', 
                  '10 to 19 employees', 
                  '20 to 99 employees', 
                  '100 to 499 employees', 
                  '500 to 999 employees', 
                  '1,000 to 4,999 employees', 
                  '5,000 to 9,999 employees', 
                  '10,000 or more employees', 
                  'I don’t know']
devtype_orgsize['OrgSize'] = pd.Categorical(devtype_orgsize['OrgSize'], categories=org_size_order, ordered=True)
df_sorted = devtype_orgsize.sort_values(by='OrgSize')

df_sorted.to_csv('devtype_orgsize.csv')
```
